```python
from utils.task1_utils import mark_and_record, files_filter, map_to_normal_distribution, draw_raw_score, draw_mapped_score, save_to_csv
from os import listdir
from os.path import isfile, join

#1: The directory where all the student results are stored
student_resutls_dir = 'gradebook_I3132-COMP-34711-1231-1SE-041039_34711-classification-S-text-representation-classification_2024-08-03-22-10-37'

#2: The gold standard file
gold_standard_file = 'gold_standards/Task-1-test-dataset1-GOLD-STANDARD.csv'

#3: Filter the student files
student_files_list = [join(student_resutls_dir, f) for f in listdir(student_resutls_dir) if isfile(join(student_resutls_dir, f))]
task1_method_a_student_files = files_filter(student_files_list, "Task1-method-a")
task1_method_b_student_files = files_filter(student_files_list, "Task1-method-b")
```

# Task1 Method a

### Raw scores

```python
task1_method_a_results = []
for path in task1_method_a_student_files:
    results = mark_and_record(path, gold_standard_file, "Task1-method-a")
    task1_method_a_results.append(results)
all_scores = [dict['total_score'] for dict in task1_method_a_results]
draw_raw_score(all_scores, "Task1-method-a")
```

### Map to a normal distribution

```python
#********************************
target_mean = 70
target_std_dev_scale = 3 # smaller value means higher standard deviation
target_max_score = 100
target_min_score = 40
#********************************

mapped_scores, mapped_results = map_to_normal_distribution(task1_method_a_results, mean=target_mean, min_value=target_min_score, max_value=target_max_score, std_dev_scale=target_std_dev_scale)
draw_mapped_score(mapped_scores)
save_to_csv(mapped_results, 'marking_results/task1_method_a_mapped_results.csv')
```
