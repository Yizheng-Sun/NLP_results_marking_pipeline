```python
from utils.task1_utils import files_filter, save_to_csv, draw_raw_score, draw_mapped_score
from utils.task2_utils import mark_and_record, map_to_normal_distribution
from os import listdir
from os.path import isfile, join

#1: The directory where all the student results are stored
student_resutls_dir = 'gradebook_I3132-COMP-34711-1231-1SE-041039_34711-classification-S-text-representation-classification_2024-08-03-22-10-37'

#2: The gold standard file
gold_standard_file = 'gold_standards/Task-2-test-dataset1-GOLD-STANDARD.csv'

#3: Filter the student files
student_files_list = [join(student_resutls_dir, f) for f in listdir(student_resutls_dir) if isfile(join(student_resutls_dir, f))]
task2_method_a_student_files = files_filter(student_files_list, "Task2-method-a")
task2_method_b_student_files = files_filter(student_files_list, "Task2-method-b")
```

# Task2 Method a

```python
all_task2_method_a_results = []
for student_file in task2_method_a_student_files:
    result = mark_and_record(student_file, gold_standard_file, "Task2-method-a")
    all_task2_method_a_results.append(result)
```

### Raw scores

```python
all_task2_method_b_f1_scores = [dict['f1_score']*100 for dict in all_task2_method_a_results]
draw_raw_score(all_task2_method_b_f1_scores, 'Task2-method-a')
```

### Map to a normal distribution

```python
#********************************
target_mean = 70
target_std_dev_scale = 3 # smaller value means higher standard deviation
target_max_score = 100
target_min_score = 40
#********************************

mapped_scores, mapped_results = map_to_normal_distribution(all_task2_method_a_results, mean=target_mean, min_value=target_min_score, max_value=target_max_score, std_dev_scale=target_std_dev_scale)
draw_mapped_score(mapped_scores)
save_to_csv(mapped_results, 'marking_results/task2_method_a_mapped_results.csv')
```
